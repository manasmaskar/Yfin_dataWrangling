<h3> Fethc Apple Data for test </h1>

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
ticker_symbol = 'AAPL'

# Retrieve historical data using yfinance
apple_data = yf.download(ticker_symbol, start="2005-01-01", end="2024-04-11")

# Calculate short-term EMA
short_term_period = 12
short_ema = apple_data['Adj Close'].ewm(span=short_term_period, min_periods=1).mean()

# Calculate long-term EMA
long_term_period = 26
long_ema = apple_data['Adj Close'].ewm(span=long_term_period, min_periods=1).mean()

# Calculate MACD line
macd_line = short_ema - long_ema

# Calculate signal line (9-period EMA of MACD line)
signal_period = 9
signal_line = macd_line.ewm(span=signal_period, min_periods=1).mean()

# Add MACD and signal line to the DataFrame
apple_data['MACD'] = macd_line
apple_data['Signal Line'] = signal_line

# Print the first few rows of the data with MACD and signal line
print(apple_data.head())
```

```python
import pandas as pd
import json

# Assuming 'df' is your DataFrame which is already loaded with data

# Convert the DataFrame to the desired JSON structure
result = {}
for name, group in combined_df.groupby('company_name'):
    # Convert date to string in ISO format (or any other format you prefer)
    group['Date'] = group['Date'].dt.strftime('%Y-%m-%d')
    # Organize the data by date within each company's data
    group_data = group.drop(columns='company_name').set_index('Date')
    result[name] = group_data.to_dict(orient='index')

# Convert to JSON string (optional, only if you need to serialize it to a JSON format string)
json_result = json.dumps(result, indent=4)

# To see the result in the console or to debug
print(json_result)

# If you need to write the JSON to a file
with open('json_CombinedStocks.json', 'w') as f:
    json.dump(result, f, indent=4)
```

<h5> Now Plotting the Stagnancy or Growth of a stock<br>
<h6> This will be plotted on the basis of calculating the rolling average <br>
<h6> Plotting the volume and rolling average<br>
<h6> Then check for any different pattern <br>
<h6> And finally, compare the Price movement 

```python

# Calculate rolling average of trading volume
window_size = 30  # You can adjust the window size as needed
rolling_avg_volume = timeSeries_df.groupby('company_name')['Volume'].rolling(window=window_size).mean()

# Plot volume and rolling average
plt.figure(figsize=(10, 6))
for company_name, group in timeSeries_df.groupby('company_name'):
    plt.figure(figsize=(10, 6))
    rolling_avg_volume = group['Volume'].rolling(window=window_size).mean()
    plt.plot(group['Date'], group['Volume'], label='Volume')
    plt.plot(group['Date'], rolling_avg_volume, label='Rolling Average Volume', linestyle='--')
    plt.title(f'Trading Volume and Rolling Average Volume Over Time for {company_name}')
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.legend()
    plt.grid(True)
    plt.show()
```
